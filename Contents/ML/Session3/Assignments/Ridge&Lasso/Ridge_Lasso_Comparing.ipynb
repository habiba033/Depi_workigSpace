```python
"""
This project demonstrates the difference between Ridge and Lasso Regression.

- Ridge Regression (L2 penalty): 
Adds the square of the coefficients to the loss function.
It shrinks coefficients but does not make them exactly zero.
Useful when we believe all features contribute a little bit.

- Lasso Regression (L1 penalty): 
Adds the absolute value of the coefficients to the loss function.
It can shrink some coefficients to exactly zero (feature selection).
Useful when we expect that only some features are truly important.

Experiment:
- By increasing the value of alpha (λ), we increase the penalty:
-> Coefficients shrink more.
-> Model becomes simpler, reducing overfitting but increasing bias.
"""

import pandas as pd
import matplotlib.pyplot as plt
import os
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge,Lasso

dataset = pd.read_csv(r'datadata.txt')  

df = dataset.copy()
X = df.iloc[: ,:-1]
y = df.iloc[: , -1]

#increasing the ridge value 
l =50
Ridgee = Ridge(alpha = l)
Ridgee.fit(X,y)
ridge_coeff = Ridgee.coef_
ridge_intercept = Ridgee.intercept_


X_test = [1,2,3,4]
y_pred = Ridgee.predict(pd.DataFrame(X_test))

plt.figure(figsize=(12,6) , dpi = 90)
plt.plot(X_test,y_pred , c ='r' , linewidth = 2)
plt.scatter(X,y)
plt.ylim(ymin = 0 , ymax = 9)
plt.xlim(xmin = 0 , xmax = 7)

# Add a text annotation to the last point of the regression line
plt.text(X_test[-1],y_pred[-1] # Position of text → last X point and its predicted Y
         , ' y = '   +    str('%.2f' %ridge_coeff)  + ' * X +'  +  str('%.2f' %ridge_intercept)    
                                                    +  '  For \u03BB or  \u03b1 = ' +    str(l)  ,   fontsize = 12)
plt.title('ridge Regression')
plt.show()
```

```python
l =1
Lassoo = Lasso(alpha = l)
Lassoo.fit(X,y)
Lasso_coeff = Lassoo.coef_
Lasso_intercept = Lassoo.intercept_


X_test = [1,2,3,4]
y_pred = Lassoo.predict(pd.DataFrame(X_test))

plt.figure(figsize=(12,6) , dpi = 90)
plt.plot(X_test,y_pred , c ='r' , linewidth = 2)
plt.scatter(X,y)
plt.ylim(ymin = 0 , ymax = 9)
plt.xlim(xmin = 0 , xmax = 7)

# Add a text annotation to the last point of the regression line
plt.text(X_test[-1],y_pred[-1] # Position of text → last X point and its predicted Y
         , ' y = '   +    str('%.2f' %Lasso_coeff)  + ' * X +'  +  str('%.2f' %Lasso_intercept)    
                                                    +  '  For \u03BB or  \u03b1 = ' +    str(l)  ,   fontsize = 12)
plt.title('Lasso Regression')
plt.show()
```
