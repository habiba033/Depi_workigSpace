```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array ([1,2,3,4])
y = np.array([2,2.8,3.6,4.5])
```

```python
w = 0 
b = 0
alpha = 0.05
num_iteration = 50
```

```python
SSE_values = []
for i in range(num_iteration):
    y_hat = (w * x) + b
    D_w = (2/len(x)) * np.sum((y_hat - y) * x)
    D_b = (2/len(x)) * np.sum(y_hat - y)
    w -= alpha * D_w
    b -= alpha * D_b
    
    SSE = (np.sum((y_hat - y)**2))
    SSE_values.append(SSE)
    
    if (i+1) % 20 == 0:
        print(f"iteration {i+1} , SSE : {SSE}")
```

```python
print(f"Optimized parameters -> slope : {w} , bais : {b}")
```

```python
plt.figure(figsize=(12,5))
plt.plot(range(num_iteration) , SSE_values , label = 'SSE')
```
