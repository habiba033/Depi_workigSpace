```python
'''
Lets you store names and phone numbers in a dictionary.
Prevents duplicates using a set.
Loops to let you add multiple contacts.
Saves them to a file so they dont disappear when program ends.
Reads them back at the start.'''

import os
contacts = {}
uniq_phones = set()
path = r"C:\Users\habib\OneDrive\المستندات\DEPI_RECAP\DEPI_RECAP\.env\python\A_Introduction To Python\S_2\path.txt"
if os.path.exists(path):
    with open(path,'r') as f:
        for line in f:
            # 1- line = name + phone 
            # line.strip -> remove spaces and \n -> "habiba|0123456789\n" -> "habiba|0123456789"
            # .split'|' -> split at | ->   "habiba|0123456789" -> [habiba , 0123456789]
            # 2- name ,phone = (means UNPACKING) -> [habiba , 0123456789] this list gonna unpack into two variables -> name ="habiba" , phone="0123456789"
            name , phone = line.strip().split('|')
            # stores data in the dict - > contacts = {'habiba':'0123456789'}
            contacts[name] = phone
            # add the phone number into a set uniq_phones -> doesn't allow duplicate
            uniq_phones.add(phone)
            
print("Type 'stop' to finish.\n")

#### ADD CONTACTS
while True : 
    name = input("enter name :")
    # if u write stop the loop while stop !
    if name.lower() == "stop":
        break
    
    #don't put int to the input -> it ignores 0 if the number start with
    phone = input("enter the phone :")
    if phone in uniq_phones:
        print("this phone number is repeated , try another!")
        # don't forget to continue
        continue   
    
    # contacts[habiba] = 0123456789
    contacts[name] = phone
    #don't foreget to add that inserted phone to uniq_phones to prevent repeating
    uniq_phones.add(phone)
    print(f"Saved : {name} --> {phone}")
    

### SHOW CONTACTS 
print("\n ALL CONTACTS :")
#.items() gives you all key–value pairs from the dictionary as tuples
for name , phone in contacts.items():
    print(name, ":", phone)
    
    
### save in file:
with open(path , 'w') as f:
    for name , phone in contacts.items():
        f.write(f"{name}|{phone}\n")
print(f"\n {len(contacts)} Contacts saved to file.")

# ''' WHAT I'VE NOTICE AFTER RUNNING 
# when u enter the same name when running again , it will update the phone number
# Full Cycle
# Start: Program checks file → if exists, load it (with split("|")).
# Run: You add new names and phones → stored in dict + set.
# End: Program saves them back to file in the same format.
# Next time you run → it picks them up again.
# '''
```
